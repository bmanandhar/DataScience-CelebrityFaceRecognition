```python
#import library
import numpy as np
import cv2
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore', '.*do not.*',)
image_path = 'http://t3.gstatic.com/licensed-image?q=tbn:ANd9GcQ2Yt_ER0yxQNhXgcmDQw_LIgpagFiXZT40BsFQxy3PlcKCFmnd_5AhiVIf4ihBGUFjX8d1qPylGXsLgiE'
```

```python
#read image, `imread`
# image1 = cv2.imread('temple.jpeg')
image = cv2.imread('./Images/road.png')
```

```python
#display image
from matplotlib import pyplot as plt
%matplotlib inline
plt.imshow(image) 
```

```python
#extract height n width of an image
h, w = image.shape[:2]
f'Height: {h} PPI, Width: {w} PPI'  
```

```python
#extract RGB vals (opencv arranges order- BGR)
B, G, R = image[100, 100]
f'R: {R}, G: {G}, B: {B}'
```

```python
#value for specific channel
B = image[100, 100, 0] # for B
G = image[100, 100, 1] # for G
R = image[100, 100, 2] # for R

f'R: {R}, G: {G}, B: {B}'
```

```python
#region of interest (roi)
#by slicing the pixels of image

roi = image[100: 500, 200: 700]
roi
```

```python
plt.imshow(roi)
```

```python
#resize, it takes 2 params, image and dimensions
#aspect ratio is not maintained, needs some extra work
resize = cv2.resize(image, (800, 800))
plt.imshow(resize)
```

```python
#calculate ratio
ratio = 800 / w
```

```python
#create a tuple (width, height)
dim = (800, int(h * ratio))
#resized image
resize_aspect = cv2.resize(image, dim)
plt.imshow(resize_aspect)
```

```python
#Rotate image
#calculate center of image
center = (w // 2, h // 2)

#generate rotation matrix
matrix = cv2.getRotationMatrix2D(center, -45, 1.0)

#Performing the affine transformation
rotated = cv2.warpAffine(image, matrix, (w, h))

plt.imshow(rotated)
```

```python
#draw a rectangle
output = image.copy()
#params: (image, (top-left corner co-ordinates), (bottom-right corner co-ordinates), (color in bgr format), line-width
rectangle = cv2.rectangle(output, (1500, 900), (600, 400), (255, 0, 0), 2)
plt.imshow(rectangle)
```

```python
#display text
output = image.copy()
text = cv2.putText(output, 'OpenCV Demo', (700, 600), cv2.FONT_HERSHEY_SIMPLEX, 4, (255, 255, 255), 6)
plt.imshow(text)
```

```python
#cvtColor() Method
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```
